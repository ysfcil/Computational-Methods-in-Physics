```python
from numpy.fft import fft, ifft

sr = 1000
ts = 1.0/sr
t = np.arange(0,1,ts)
data=pd.read_table(r"C:\Users\yusuf\Downloads\3_b_5ms_int_1200 lmm_1.asc (1).txt")
	
data.columns = ["0", "1"]
data.plot(x='0', y='1', kind='line') 
plt.show()
X = data["1"]
X = fft(X)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.show()
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import numpy as np

peaks, _ = find_peaks(np.abs(X), distance=150)
# difference between peaks is >= 150
print(np.diff(peaks))
# prints [186 180 177 171 177 169 167 164 158 162 172]
plt.plot(X)
plt.plot(peaks, X[peaks], "x")
plt.plot(np.zeros_like(X), "--", color="gray")
plt.show()
```
