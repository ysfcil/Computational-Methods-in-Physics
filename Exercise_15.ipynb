```python
import numpy as np
import matplotlib.pyplot as plt
import time
import random
```

```python
decay = 0.0008675
N = 100
N_step = [N] * 10000
while True:
    x=random.random()
    if x<decay:
        N-=1
    if N == 0:
        break
    N_step.append(N)
plt.plot(range(len(N_step)), N_step)
    
```
