```python
#090200118 Yusuf Çil
#Question 2 MonteCarlo Integral
#Code:
a = -1
b = 1 
N = 1000
ar = np.zeros(N)
for i in range (len(ar)):
    ar[i] = random.uniform(a,b)
integral = 0.0
def f(x):
    return x**2
for i in ar:
    integral += f(i)

ans = ((b-a)/float(N))*integral
print ("The value is {}.".format(ans))


```

```python
#090200118 Yusuf Çil
#Question3 runge kutta 
def f (z,h,u):
    return u
def g(z,h,u):
    return -b*u-(w*w)*h - np.sin(z)
w=pow(5, 1/2)
b=3
def f (z,h,u):
    return u
def g(z,h,u):
    return -b*u-(w*w)*h
znitial=0
hinitial=1
uinitial=3
zfinal=7
numpoints=1000
q=(zfinal-znitial)/numpoints

zvalues = np.linspace(0,7, 1000)
hvalues = np.zeros(numpoints)
hvalues[0] = hinitial
uvalues = np.zeros(1000)
uvalues[0] = uinitial
for i in range(numpoints-1):
    z = zvalues[i]
    h = hvalues[i]
    u = uvalues[i]
    F1h = f(z,h,u)
    F1u = g(z,h,u)
    F2h=f (z+(q/2 ) , z+(q/2 ) *F1h , u+(q/2 ) *F1u )
    F2u=g (z+(q/2 ) , z+(q/2 ) *F1h , u+(q/2 ) *F1u )
    F3h=f (z+(q/2 ) , z+(q/2 ) *F2h , u+(q/2 ) *F2u )
    F3u=g (z+(q/2 ) , z+(q/2 ) *F2h , u+(q/2 ) *F2u )
    F4h = f(z+q, z + q*F3h, u+q*F3u)
    F4u = f(z+q, z + q*F3h, u+q*F3u)
    hvalues[i+1] = hvalues[i]+(q/6)*(F1h+2*F2h+2*F3h+F4h)
    uvalues[i+1] = uvalues[i]+(q/6)*(F1u+2*F2u+2*F3u+F4u)
print(hvalues[-1])
plt.plot(zvalues,hvalues)

```

```python
#090200118 Yusuf Çil
#QUESTİON 4 plot fit
M = np.zeros((8, 8))
for i in range(0,8):
  for j in range(0,8):
    sum_ = 0
    for k in range(1000):
      sum_ += zvalues[k]**(i+j)
    M[i][j] = sum_
    M[0][0] = 7
v = np.zeros((8))
for i in range(8):
    sum_ = 0
    for k in range(1000):
      sum_ += zvalues[k]**i * hvalues[k]
    v[i] = sum_
A = np.linalg.solve(M,v)
print(A)
p1 = np.poly1d(np.flip(A))
print(p1)
h_fit_values = np.zeros((1000))
for i in range(1000):
  h_fit_values[i] = p1(zvalues[i])
plt.plot(zvalues, h_fit_values)
      



```
