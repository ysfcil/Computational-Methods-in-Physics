```python
import numpy as np
import matplotlib.pyplot as plt
import time
import random
```

```python
distance_from_home = 50
dfhlist = []
while True:
 chance_ = random.random()
 if chance_ >= 0.5:
    distance_from_home += 1
 else:
    distance_from_home -=1
 dfhlist.append(distance_from_home)
 if distance_from_home == 0 or distance_from_home == 100:
   break
plt.plot(range(len(dfhlist)), dfhlist)
```
