```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import random as random
import time
timeValues_for = []
timeValues_vector = []
nvalues = [1,2,3,4,5,6]
```

```python
for k in range(0, 6):  
 x = np.random.rand(10**(k+1))
 my_sum = 0.0

 tic = time.time()
 for i in range(len(x)):
    if x[i] > 0.5:
     my_sum += x[i]
 toc = time.time()
 timeValues_for.append(toc-tic)

 filter_arr = x > 0.5
 tic = time.time()
 new_x = x[filter_arr]
 a = new_x.sum()
 toc = time.time()
 timeValues_vector.append(toc-tic)
```

```python
plt.plot(nvalues,timeValues_for)
plt.gca().twinx().plot(nvalues, timeValues_vector, color = 'r')
plt.ylim((0,0.200))
plt.show
```
