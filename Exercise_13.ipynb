```python
import numpy as np
import matplotlib.pyplot as plt
import time
import random
```

```python
hidrojen = 10000
oxygen = 10000
h2o = 0
numberof_ox = []
numberof_hid = []

while True:
    chance_ = random.random()
    if (hidrojen/oxygen) > chance_:
        oxygen -= 1
        hidrojen -= 2
        h2o += 1
       
    numberof_ox.append(oxygen)
    numberof_hid.append(hidrojen)
    if hidrojen == 0:
        break
```

```python
plt.plot(range(len(numberof_ox)),numberof_ox)
plt.gca().twinx().plot(range(len(numberof_ox)), numberof_hid, color = 'r')
plt.ylim((0,10000))
plt.show


plt.show
```
