```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
v1 = []
v2 = np.zeros(10**6)
powers = [0,1,2,3,4,5,6]
with_preallocation = []
no_preallocation = []
for k in powers:

 tic = time.time()
 for i in range(10**k):
    v1.append(i*i)
 toc = time.time()
 no_preallocation.append(toc - tic)

 tic = time.time()
 for i in range(0, 10**k - 1):
    v2[i] = i*i

 toc = time.time()
 with_preallocation.append(toc-tic)

plt.plot(powers,no_preallocation)
plt.gca().twinx().plot(powers, with_preallocation, color = 'r')
plt.ylim((0,0.900))
plt.show
```
